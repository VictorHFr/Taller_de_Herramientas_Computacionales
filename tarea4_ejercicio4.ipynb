```python
import numpy as np
import matplotlib.pyplot as plt

a, b = 0, 3  # Límites
M = 10.0428  # Cota superior
n_values = np.logspace(1, 6, 100).astype(int)  # Valores de n
I_estimates = []  # Lista estimaciones

# Valor real de la integral
I_real = (np.exp(b) - np.exp(a)) / 2

for n in n_values:
    x = np.random.uniform(a, b, n)  # Generar x en [0, 3]
    y = np.random.uniform(0, M, n)  # Generar y en [0, M]
    fav = np.sum(y <= np.exp(x) / 2)  # Contar casos favorables
    I_estimate = (fav / n) * (b - a) * M  # Estimación de la integral
    I_estimates.append(I_estimate)

# Gráfica de convergencia
plt.figure(figsize=(10, 6))
plt.plot(n_values, I_estimates, label="Estimación Monte Carlo", color="blue")
plt.axhline(I_real, color="red", linestyle="--", label="Valor real")
plt.xscale("log")
plt.xlabel("Número de lanzamientos (n)")
plt.ylabel("Estimación de la integral")
plt.title("Convergencia del estimador Monte Carlo")
plt.legend()
plt.grid()
plt.show()

# Estimación del error
error = np.abs(np.array(I_estimates) - I_real)
plt.figure(figsize=(10, 6))
plt.plot(n_values, error, label="Error", color="green")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Número de lanzamientos (n)")
plt.ylabel("Error absoluto")
plt.title("Error de la estimación Monte Carlo")
plt.legend()
plt.grid()
plt.show()
```

```python
a, b = 0, 3  # Límites de integración
M = 10.0428  # Cota superior
n = 1000  # Número de lanzamientos (puntos aleatorios)

# Valor real de la integral
I_real = (np.exp(b) - np.exp(a)) / 2


x = np.random.uniform(a, b, n)  # Generar x en [0, 3]
y = np.random.uniform(0, M, n)  # Generar y en [0, M]
fav = y <= np.exp(x) / 2  # Puntos favorables (bajo la curva)
I_estimate = (np.sum(fav) / n) * (b - a) * M  # Estimación de la integral

# Gráfica de los puntos generados
plt.figure(figsize=(10, 6))
plt.scatter(x[fav], y[fav], color="blue", label="Puntos bajo la curva", alpha=0.6)
plt.scatter(x[~fav], y[~fav], color="red", label="Puntos fuera de la curva", alpha=0.6)

# Graficar la función f(t) = e^t / 2
t = np.linspace(a, b, 500)
plt.plot(t, np.exp(t) / 2, color="black", label=r"$f(t) = \frac{e^t}{2}$", linewidth=2)

# Añadir detalles a la gráfica
plt.xlabel("t")
plt.ylabel("f(t)")
plt.title(f"Método de Monte Carlo para estimar la integral (n = {n})")
plt.legend()
plt.grid()
plt.show()

# Resultado de la estimación
print(f"Estimación de la integral: {I_estimate}")
print(f"Valor real de la integral: {I_real}")
print(f"Error absoluto: {np.abs(I_estimate - I_real)}")
```
