1.- Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si es primo o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
def es_primo(numero):
    if numero <= 1:
        return "Error en el Número"

    for i in range(2, int(numero**0.5) + 1):
        if numero % i == 0:
            return False

    return True

def main():
    while True:
        try:
            num = int(input("Ingrese un número entero mayor que 1: "))
            if num > 1:
                break
            else:
                print("Error: El número debe ser mayor que 1.")
        except ValueError:
            print("Error: Ingrese un número entero válido.")

    resultado = es_primo(num)

    print(f"El número {num} es primo: {resultado}")

if __name__ == "__main__":
    main()
```

2.- Sin utilizar librerías adicionales cree un arreglo de numéros enteros (entre 0 y 150) y calcule a través de funciones estándar o lambda lo siguiente:

El promedio

La varianza

La desviación estándar

Visualice los datos en un histograma.

La suma de los números pares e impares.

Los múltiplos de 2, 3 y 5.

```python
import random


# Crear un arreglo de números enteros entre 0 y 150
datos=[random.randint(0, 150) for _ in range(50)]

# Funciones para calcular el promedio, la varianza y la desviación estándar
calcular_promedio = lambda arr: sum(arr) / len(arr) if len(arr) > 0 else 0

def calcular_varianza(arr, promedio):
    return sum((x - promedio) ** 2 for x in arr) / len(arr) if len(arr) > 0 else 0

def calcular_desviacion_estandar(varianza):
    return varianza ** 0.5

# Funciones para separar números pares e impares y calcular sus sumas
separar_pares_impares = lambda arr: ([num for num in arr if num % 2 == 0], [num for num in arr if num % 2 != 0])

calcular_suma = lambda arr: sum(arr)

# Funciones para encontrar múltiplos de 2, 3 y 5
encontrar_multiplos = lambda arr, divisor: [num for num in arr if num % divisor == 0]

import matplotlib.pyplot as plt
# Visualizar los datos en un histograma
plt.hist(datos, bins=30, color='orange', edgecolor='black')
plt.title('Histograma de Números Aleatorios')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()


# Calcular promedio, varianza y desviación estándar
promedio = calcular_promedio(datos)
varianza = calcular_varianza(datos, promedio)
desviacion_estandar = calcular_desviacion_estandar(varianza)

# Separar números pares e impares y calcular sus sumas
pares, impares = separar_pares_impares(datos)
suma_pares = calcular_suma(pares)
suma_impares = calcular_suma(impares)

# Encontrar múltiplos de 2, 3 y 5
multiplos_2 = encontrar_multiplos(datos, 2)
multiplos_3 = encontrar_multiplos(datos, 3)
multiplos_5 = encontrar_multiplos(datos, 5)


# Imprimir resultados
print("el arreglo numerico es:",datos)
print(f"Promedio: {promedio}")
print(f"Varianza: {varianza}")
print(f"Desviación Estándar: {desviacion_estandar}")
print(f"Suma de Pares: {suma_pares}")
print(f"Suma de Impares: {suma_impares}")
print(f"Múltiplos de 2: {multiplos_2}")
print(f"Múltiplos de 3: {multiplos_3}")
print(f"Múltiplos de 5: {multiplos_5}")
```

3.- En la clase de ecuaciones diferenciales se pide a los estudiantes encontrar la solución del problema de valor inicial:

\begin{equation}
y´=-y+x \hspace{3cm} y(-1)=5.
\end{equation}


Los estudiantes llegaron a la solución general cuya ecuación es:

\begin{equation}
y=(x-1)+C e^{-x}
\end{equation}

Apoye a sus compañeros y grafique en python la familia de curvas para la solución de la ecuación diferencial y en un sólo gráfico muestre de diferente color las curvas para los siguientes tres casos:

*   para valores de C > 0
*   cuando C = 0
*   para valores de C < 0

Señale en el gráfico las curvas que corresponden a cada uno de los tres casos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función que representa la solución general
def solucion_general(x, C):
    return (x - 1) + C * np.exp(-x)

# Crear un rango de valores para x
x_vals = np.linspace(-3, 3, 400)

# Graficar las curvas para diferentes valores de C
for C in np.linspace(-1, 1, 100):
    y_vals = solucion_general(x_vals, C)
    if C < 0:
        plt.plot(x_vals, y_vals, color='blue')
    elif C == 0:
        plt.plot(x_vals, y_vals, color='green')
    else:
        plt.plot(x_vals, y_vals, color='red')

# Configuración del gráfico
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Familia de Curvas para la Ecuación Diferencial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Mostrar el gráfico
plt.show()
```

4.- Sea la función:
\begin{equation}
\frac{3x + 7}{x^2 - 4}
\end{equation}

Utilizando la librería matplotlib:

1.- Grafique la función con línea continua para un rango de -10 a 10, con más de 1000 muestras.

2.- Investigue en la literatura cómo graficar las asíntotas horizontales y verticales. Muéstrelas en el gráfico con líneas punteadas.

3.- Establezca límites en el eje "y" para evitar que la gráfica muestre valores muy altos o muy bajos cerca de las asíntotas verticales.

4.- Añada una leyenda indicando los elementos presentes en la gráfica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def funcion(x):
    return (3*x + 7) / (x**2 - 4)

# Rango de valores
x_vals = np.linspace(-10, 10, 1001)

# Graficar la función
plt.plot(x_vals, funcion(x_vals), label='Función', color='blue')

# asíntotas horizontales y verticales
plt.axhline(0, color='black', linestyle='--', label='Asíntota Horizontal')
plt.axvline(-2, color='red', linestyle='--', label='Asíntota Vertical')
plt.axvline(2, color='red', linestyle='--', label='Asíntota Vertical')

# Establecer límites en el eje "y"
plt.ylim(-10, 10)
plt.legend()

# Configuración del gráfico
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Gráfico de la Función con Asíntotas')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar el gráfico
plt.show()
```
